```python
df_main_notna = df_main[df_main['period'].notna()]
period_by_subsector = df_main_notna.groupby('ssector').median()[['period']].to_dict()
```

```python
period_by_sector = df_main_notna.groupby('sector').median()[['period']].to_dict()
mydict_period = period_by_subsector['period']
df_main_imputed = df_main
df_main_imputed.period = df_main.period.fillna(df_main.ssector.map(mydict_period))
```

```python
df_main_imputed['End Year'] = df_main_imputed['FCY']+df_main_imputed['period']
```

```python
df_main_imputed['range'] = df_main_imputed.apply(lambda row: np.arange(row['FCY'], np.nanmin([row['FCY']+row['period']+1,2020,row['DSU Year']+1])), axis=1)
```

```python
df_main_imputed['status_n'].drop_duplicates()
```

```python
def new_range(row):
    if row['status_n']=='Active':
        out = np.arange(row['FCY'], np.nanmin([row['FCY']+row['period']+1,2020]))
    elif row['status_n'] == 'Distressed':
        out = np.arange(row['FCY'], np.nanmin([row['FCY']+row['period']+1,2020]))
    elif row['status_n'] == 'Concluded':
        out = np.arange(row['FCY'], np.nanmin([row['FCY']+row['period']+1,2020]))
    elif row['status_n'] == 'Cancelled':
        out = np.arange(row['FCY'], np.nanmin([row['FCY']+row['period']+1,2020,row['DSU Year']+1]))
    else:
        out = np.arange(row['FCY'], np.nanmin([row['FCY']+row['period']+1,2020]))
    return out
```

```python
df_main_imputed['range2'] = df_main_imputed.apply(lambda row: new_range(row) , axis=1)
```
